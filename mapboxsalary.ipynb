```python
import os
from mapboxgl.viz import *
from mapboxgl.utils import *
```

```python
# Must be a public token, starting with `pk`
token = ('public token')
```

```python
# create choropleth from polygon features stored as GeoJSON
viz6 = ChoroplethViz('https://raw.githubusercontent.com/secondteacher/Malaysia-Polygon-Map-with-State-geojson-file/master/salary2018.geojson', 
                    access_token=token,
                    color_property='Median Salary',
                    color_stops=create_color_stops([3000, 3250, 3500, 4000, 4500], colors='YlOrRd'),
                    color_function_type='interpolate',
                    line_stroke='--',
                    line_color='rgb(128,0,38)',
                    line_width=1,
                    line_opacity=0.9,
                    opacity=0.8,
                    center=(102, 4.2),
                    zoom=3,
                    below_layer='waterway-label',
                    legend_layout='horizontal',
                    legend_key_shape='bar',
                    legend_key_borders_on=False)
viz6.show()
```

```python
# Map settings
viz6.style='mapbox://styles/mapbox/dark-v9?optimize=true'
viz6.label_color = 'hsl(0, 0%, 70%)'
viz6.label_halo_color = 'hsla(0, 0%, 10%, 0.75)'
viz6.height = '400px'

# Legend settings
viz6.legend_gradient = False
viz6.legend_fill = '#343332'
viz6.legend_header_fill = '#343332'
viz6.legend_text_color = 'hsl(0, 0%, 70%)'
viz6.legend_key_borders_on = False
viz6.legend_title_halo_color = 'hsla(0, 0%, 10%, 0.75)'

# Render map
viz6.show()
```

```python
with open('viz6.html', 'w') as f:
    f.write(viz.create_html())
```

```python
viz6.add_snapshot_links = True
viz6.show()
```

```python
# adjust view angle
viz6.bearing = -14
viz6.pitch = 45

# add extrusion to viz using interpolation keyed on density in GeoJSON features
viz6.height_property = 'Median Salary'
viz6.height_stops = create_numeric_stops([0,3000, 3250, 3500, 4000, 4500, 5000], 0, 100000)
viz6.height_function_type = 'interpolate'

# render again
viz6.show()
```
